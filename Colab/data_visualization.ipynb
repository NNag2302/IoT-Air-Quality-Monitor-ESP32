```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load CSV dataset
df = pd.read_csv("dataset.csv", parse_dates=["Timestamp"])

plt.figure(figsize=(14, 6))

# Plot each MQ sensor
plt.plot(df["Timestamp"], df["MQ2"], label="MQ2 (ppm)", linewidth=2)
plt.plot(df["Timestamp"], df["MQ4"], label="MQ4 (ppm)", linewidth=2)
plt.plot(df["Timestamp"], df["MQ135"], label="MQ135 (ppm)", linewidth=2)

plt.title("MQ Sensor Readings Over Time", fontsize=16)
plt.xlabel("Timestamp")
plt.ylabel("Sensor Values (ppm)")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%H:%M:%S"))
plt.show()
```
